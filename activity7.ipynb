```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import folium
from folium.plugins import HeatMap
warnings.filterwarnings('ignore')

```

```python
df = pd.read_csv( 'data-sets//Global_terrorist.csv', encoding='ISO-8859-1')
```

```python
df.columns
```

<h1>Filtering Data</h1>

```python
#regional

filter1 = df[df['region_txt'] == 'Middle East & North Africa']
filter2 = filter1[filter1['attacktype1_txt'] == 'Bombing/Explosion']
filter3 = filter2[filter2['nkill'] >= 10]
filter3
```

<h1>Aggretion</h1>

```python
aggregate_df = filter3.groupby(['latitude','longitude']).size().reset_index(name='incident_count')
aggregate_df
```

```python
heat_data = aggregate_df[['latitude','longitude','incident_count']].values.tolist()
heat_data
```

```python
 
m = folium.Map(location=[30,40],zoom_start=4)
HeatMap(heat_data).add_to(m)
m
```

<h1>Regional Attack</h1>

```python
#Region:South Asia
#Attacktype:Armed Assault
#Year:200

filter1 = df[df['region_txt'] == 'South Asia']
filter2 = filter1[filter1['attacktype1_txt'] == 'Armed Assault']
filter3 = filter2[filter2['nkill'] >= 10]
filter4 = filter3[filter3['iyear'] > 2000]
#filter5 = filter4[filter4['weaptype1_txt'] == 'Firearms']
filter4
```

```python
aggregate_df = filter4.groupby(['latitude','longitude']).size().reset_index(name='incident_count')
heat_data = aggregate_df[['latitude','longitude','incident_count']].values.tolist()
heat_data
```

```python
 
m = folium.Map(location=[30,40],zoom_start=4)
HeatMap(heat_data).add_to(m)
m
```
